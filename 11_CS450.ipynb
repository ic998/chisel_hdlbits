## Cs450/timer

refer to https://hdlbits.01xz.net/wiki/Cs450/timer

```scala
class Timer extends RawModule {
    val clk = IO(Input(Clock()))
    val load = IO(Input(UInt(1.W)))
    val data = IO(Input(UInt(10.W)))
    val tc = IO(Output(UInt(1.W)))
    
    val counter = withClock(clk) {
        Reg(UInt(10.W))
    }
    
    when(load===1.U) {
        counter := data
    }.elsewhen(counter===0.U) {
        counter := 0.U
    }.otherwise {
        counter := counter - 1.U
    }
    
    tc := counter===0.U
}

val vlog = (new ChiselStage).emitVerilog(new Timer)
println(vlog)
```

## Cs450/counter 2bc

refer to https://hdlbits.01xz.net/wiki/Cs450/counter_2bc

```scala
class Counter2bc extends RawModule {
    val clk = IO(Input(Clock()))
    val areset = IO(Input(AsyncReset()))
    val train_valid = IO(Input(UInt(1.W)))
    val train_taken = IO(Input(UInt(1.W)))
    val state = IO(Output(UInt(2.W)))
    
    val s_param = Enum(4)
    val SNT = s_param(0)
    val WNT = s_param(1)
    val WT = s_param(2)
    val ST = s_param(3)

    val next_state = Wire(UInt(2.W))
    val state_r = withClockAndReset(clk, areset) {
        RegInit(WNT)
    }
    state_r := next_state 

    when(state===SNT) {
        next_state := Mux(train_valid===1.U && train_taken===1.U, WNT, SNT)
    }.elsewhen(state===WNT) {
        next_state := Mux(train_valid===1.U && train_taken===1.U, WT, Mux(train_valid===1.U && train_taken===0.U, SNT, WNT))
    }.elsewhen(state===WT) {
        next_state := Mux(train_valid===1.U && train_taken===1.U, ST, Mux(train_valid===1.U && train_taken===0.U, WNT, WT))
    }.elsewhen(state===ST) {
        next_state := Mux(train_valid===1.U && train_taken===0.U, WT, ST)
    }.otherwise {
        next_state := state
    }

    state := state_r
}

val vlog = (new ChiselStage).emitVerilog(new Counter2bc)
println(vlog)
```

## history shift 

refer to https://hdlbits.01xz.net/wiki/Cs450/history_shift

```scala
class HistoryShift(width: Int=32) extends RawModule {
    val clk = IO(Input(Clock()))
    val areset = IO(Input(AsyncReset()))
    
    val train_taken = IO(Input(UInt(1.W)))
    val train_mispredicted = IO(Input(UInt(1.W)))
    val train_history = IO(Input(UInt(width.W)))
    
    val predict_valid = IO(Input(UInt(1.W)))
    val predict_taken = IO(Input(UInt(1.W)))
    
    val predict_history = IO(Output(UInt(width.W)))
    
    val predict_history_r = withClockAndReset(clk, areset) {
        RegInit(0.U(width.W))
    }

    when(train_mispredicted===1.U) {
        predict_history_r := Cat(train_history(width-2,0),train_taken)
    }.elsewhen(predict_valid===1.U) {
        predict_history_r := Cat(predict_history_r(width-2,0),predict_taken)
    }

    predict_history := predict_history_r 
}

val vlog = (new ChiselStage).emitVerilog(new HistoryShift)
println(vlog)
```

## Cs450/gshare

refer to https://hdlbits.01xz.net/wiki/Cs450/gshare

```scala
class HistoryShift_7(width: Int=7) extends RawModule {
    val clk = IO(Input(Clock()))
    val areset = IO(Input(AsyncReset()))
    
    val train_valid = IO(Input(UInt(1.W)))
    val train_taken = IO(Input(UInt(1.W)))
    val train_mispredicted = IO(Input(UInt(1.W)))
    val train_history = IO(Input(UInt(width.W)))
       
    val predict_valid = IO(Input(UInt(1.W)))
    val predict_taken = IO(Input(UInt(1.W)))
    
    val predict_history = IO(Output(UInt(width.W)))
    
    val predict_history_r = withClockAndReset(clk, areset) {
        RegInit(0.U(width.W))
    }

    when(train_mispredicted===1.U && train_valid===1.U) {
        predict_history_r := Cat(train_history(width-2,0),train_taken)
    }.elsewhen(predict_valid===1.U) {
        predict_history_r := Cat(predict_history_r(width-2,0),predict_taken)
    }

    predict_history := predict_history_r 
}


class GsharePredictor extends RawModule {
    val clk = IO(Input(Clock()))
    val areset = IO(Input(AsyncReset()))

    // 用于产生跳转地址和跳转方向
    val predict_valid = IO(Input(UInt(1.W)))
    val predict_pc = IO(Input(UInt(7.W))) 

    // 用于更新BTB和PHT
    val train_valid = IO(Input(UInt(1.W)))
    val train_taken = IO(Input(UInt(1.W)))
    val train_mispredicted = IO(Input(UInt(1.W)))
    val train_history = IO(Input(UInt(7.W)))
    val train_pc = IO(Input(UInt(7.W)))
    
    val predict_history = IO(Output(UInt(7.W))) // 用组合逻辑产生
    val predict_taken = IO(Output(UInt(1.W))) // 用组合逻辑产生

    val entry_depth = 128
    val gbh_depth = 7

    // PHT: predict时根据pc获取跳转方向；train时根据pc更新跳转方向
    val PHT = Wire(Vec(entry_depth, UInt(2.W)))
    // BTB: predict时根据pc获取跳转地址

    /* ====== 预测 ====== */
    val u_gbh = Module(new HistoryShift_7(gbh_depth))
    u_gbh.clk := clk
    u_gbh.areset := areset 
    u_gbh.train_valid := train_valid
    u_gbh.train_taken := train_taken
    u_gbh.train_mispredicted := train_mispredicted
    u_gbh.train_history := train_history
    u_gbh.predict_valid := predict_valid
    u_gbh.predict_taken := predict_taken
    predict_history := u_gbh.predict_history
    val pht_index = predict_pc ^ predict_history
    when(predict_valid===1.U){
        predict_taken := PHT(pht_index)(1)
    }.otherwise {
        predict_taken := 0.U
    }
    
    /* ====== train ====== */
    val train_index = train_pc ^ train_history
    for (i <- 0 until entry_depth) {
        val u_ptaken = Module(new Counter2bc)
        u_ptaken.clk := clk
        u_ptaken.areset := areset 
        u_ptaken.train_valid := train_valid & train_index===i.U
        u_ptaken.train_taken := train_taken
        PHT(i) := u_ptaken.state
    }
}

val vlog = (new ChiselStage).emitVerilog(new GsharePredictor)
println(vlog)
```
