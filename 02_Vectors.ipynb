# Vectors

Vectors are used to group related signals using one name to make it more convenient to manipulate. For example, wire [7:0] w; declares an 8-bit vector named w that is functionally equivalent to having 8 separate wires.

## Setup Environment 

```scala
// val scalaVersion:="2.12.10"

import $ivy.`edu.berkeley.cs::chisel3:3.4.4`
import chisel3._
import chisel3.stage.ChiselStage
import chisel3.util._
```

```scala
## Vector 0

refer to https://hdlbits.01xz.net/wiki/Vector0
```

```scala
class TopModuleIO_1 extends Bundle {
    val vec = Input(UInt(3.W))
    val outv = Output(UInt(3.W))
    val o0 = Output(UInt(1.W))
    val o1 = Output(UInt(1.W))
    val o2 = Output(UInt(1.W))
}

class TopModule_1 extends RawModule {
    val ports = new TopModuleIO_1
    val vec = IO(ports.vec)
    val outv = IO(ports.outv)
    val o0 = IO(ports.o0)
    val o1 = IO(ports.o1)
    val o2 = IO(ports.o2)

    outv := vec
    o0 := vec(0)
    o1 := vec(1)
    o2 := vec(2)
}

val vlog = (new ChiselStage).emitVerilog(new TopModule_1)
println(vlog)
```

## Vector2

A 32-bit vector can be viewed as containing 4 bytes (bits [31:24], [23:16], etc.). Build a circuit that will reverse the byte ordering of the 4-byte word.

AaaaaaaaBbbbbbbbCcccccccDddddddd => DdddddddCcccccccBbbbbbbbAaaaaaaa

```scala
class TopModuleIO_3 extends Bundle {
    val in = Input(UInt(32.W))
    val out = Output(UInt(32.W))
}

class TopModule_3 extends RawModule {
    val ports = new TopModuleIO_3
    val in = IO(ports.in)
    val out = IO(ports.out)

    out := Cat(in(7,0),in(15,8),in(23,16),in(31,24))
}

val vlog = (new ChiselStage).emitVerilog(new TopModule_3)
println(vlog)
```

## Vectorgates

refer to https://hdlbits.01xz.net/wiki/Vectorgatesmn  

```scala
class TopModuleIO_4 extends Bundle {
    val a = Input(UInt(3.W))
    val b = Input(UInt(3.W))
    val out_or_bitwise = Output(UInt(3.W))
    val out_or_logical = Output(UInt(1.W))
    val out_not = Output(UInt(6.W))
}

class TopModule_4 extends RawModule {
    val ports = new TopModuleIO_4
    val a = IO(ports.a)
    val b = IO(ports.b)
    val out_or_bitwise = IO(ports.out_or_bitwise)
    val out_or_logical = IO(ports.out_or_logical)
    val out_not = IO(ports.out_not)

    out_or_bitwise := a|b 
    out_or_logical := Cat(b,a).orR()
    out_not := ~Cat(b,a)
}

val vlog = (new ChiselStage).emitVerilog(new TopModule_4)
println(vlog)
```

## Gates4

refer to https://hdlbits.01xz.net/wiki/Gates4

```scala
class TopModuleIO_5 extends Bundle {
    val in = Input(UInt(4.W))
    val out_and = Output(UInt(1.W))
    val out_or = Output(UInt(1.W))
    val out_xor = Output(UInt(1.W))
}

class TopModule_5 extends RawModule {
    val ports = new TopModuleIO_5
    val in = IO(ports.in)
    val out_and = IO(ports.out_and)
    val out_or = IO(ports.out_or)
    val out_xor = IO(ports.out_xor)

    out_and := in.andR()
    out_or := in.asBools.reduce(_ || _)
    out_xor := in.xorR()
}

val vlog = (new ChiselStage).emitVerilog(new TopModule_5)
println(vlog)
```

## Vector3 

refer to https://hdlbits.01xz.net/wiki/Vector3

```scala
class TopModuleIO_6 extends Bundle {
    val a = Input(UInt(5.W))
    val b = Input(UInt(5.W))
    val c = Input(UInt(5.W))
    val d = Input(UInt(5.W))
    val e = Input(UInt(5.W))
    val f = Input(UInt(5.W))
    val w = Output(UInt(8.W))
    val x = Output(UInt(8.W))
    val y = Output(UInt(8.W))
    val z = Output(UInt(8.W))
}

class TopModule_6 extends RawModule {
    val ports = new TopModuleIO_6
    val a = IO(ports.a)
    val b = IO(ports.b)
    val c = IO(ports.c)
    val d = IO(ports.d)
    val e = IO(ports.e)
    val f = IO(ports.f)
    val w = IO(ports.w)
    val x = IO(ports.x)
    val y = IO(ports.y)
    val z = IO(ports.z)

    w := Cat(a,b(4,2))
    x := Cat(b(1,0),c,d(4))
    y := Cat(d(3,0),e(4,1))
    z := Cat(e(0),f,3.U)
}

val vlog = (new ChiselStage).emitVerilog(new TopModule_6)
println(vlog)
```

## Vectorr 

Given an 8-bit input vector [7:0], reverse its bit ordering.

```scala
class TopModuleIO_7 extends Bundle {
    val in = Input(UInt(8.W))
    val out = Output(UInt(8.W))
}

class TopModule_7 extends RawModule {
    val ports = new TopModuleIO_7
    val in = IO(ports.in)
    val out = IO(ports.out)

    out := Reverse(in)
}

val vlog = (new ChiselStage).emitVerilog(new TopModule_7)
println(vlog)
```

## Vector4

Build a circuit that sign-extends an 8-bit number to 32 bits. 

```scala
class TopModuleIO_8 extends Bundle {
    val in = Input(UInt(8.W))
    val out = Output(UInt(32.W))
}

class TopModule_8 extends RawModule {
    val ports = new TopModuleIO_8
    val in = IO(ports.in)
    val out = IO(ports.out)

    out := Cat(Fill(24,in(7)),in)
}

val vlog = (new ChiselStage).emitVerilog(new TopModule_8)
println(vlog)
```

## Vector 5

refer to https://hdlbits.01xz.net/wiki/Vector5

```scala
class TopModuleIO_9 extends Bundle {
    val a = Input(UInt(1.W))
    val b = Input(UInt(1.W))
    val c = Input(UInt(1.W))
    val d = Input(UInt(1.W))
    val e = Input(UInt(1.W))
    val out = Output(UInt(25.W))
}

class TopModule_9 extends RawModule {
    val ports = new TopModuleIO_9
    val a = IO(ports.a)
    val b = IO(ports.b)
    val c = IO(ports.c)
    val d = IO(ports.d)
    val e = IO(ports.e)
    val out = IO(ports.out)

    val t1 = Cat(Fill(5,a), Fill(5,b), Fill(5,c), Fill(5,d), Fill(5,e))
    val t2 = Fill(5,Cat(a,b,c,d,e))
    out := ~(t1^t2)
}

val vlog = (new ChiselStage).emitVerilog(new TopModule_9)
println(vlog)
```
