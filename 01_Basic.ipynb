# Simple wire
## Wire
Create a module with one input and one output that behaves like a wire 

```scala
// val scalaVersion:="2.12.10"

import $ivy.`edu.berkeley.cs::chisel3:3.4.4`

import chisel3._
import chisel3.stage.ChiselStage
```

```scala
class TopModuleIO_1 extends Bundle {
    val in = Input(UInt(1.W))
    val out = Output(UInt(1.W))
}

class TopModule_1 extends RawModule {
    val ports = new TopModuleIO_1
    
    val in = IO(ports.in)
    val out = IO(ports.out)
    out := in
}

val vlog = (new ChiselStage).emitVerilog(new TopModule_1)
println(vlog)
//chisel3.Driver.execute(Array("--target-dir", "generated"), () => new TopModule_01_1)
```

## Wire4

Create a module with 3 inputs and 4 outputs that behaves like wires that makes these connections:
a -> w
b -> x
b -> y
c -> z

```scala
class TopModuleIO_2 extends Bundle {
    val a = Input(UInt(1.W))
    val b = Input(UInt(1.W))
    val c = Input(UInt(1.W))
    val w = Output(UInt(1.W))
    val x = Output(UInt(1.W))
    val y = Output(UInt(1.W))
    val z = Output(UInt(1.W))
}

class TopModule_2 extends RawModule{
    val ports = new TopModuleIO_2

    val a = IO(ports.a)
    val b = IO(ports.b)
    val c = IO(ports.c)
    val w = IO(ports.w)
    val x = IO(ports.x)
    val y = IO(ports.y)
    val z = IO(ports.z)

    w := a
    x := b
    y := b
    z := c
}

val vlog = (new ChiselStage).emitVerilog(new TopModule_2)
println(vlog)
```

## NorGate

Create a module that implements an NOR gate (with parameter). 

```scala
class TopModuleIO_4 (
    width: Int
) extends Bundle {
    val a = Input(UInt(width.W))
    val b = Input(UInt(width.W))
    val out = Output(UInt(width.W))
}

class TopModule_4 (
    width: Int
) extends RawModule {
    val ports = IO(new TopModuleIO_4(width))

    ports.out := ports.a | ports.b 
}

val vlog = (new ChiselStage).emitVerilog(new TopModule_4(4))
println(vlog)
```

## XnorGate & XorGate

Create a module that implements an XNOR and an XOR gate.

```scala
class TopModuleIO_5 extends Bundle {
    val a = Input(UInt(1.W))
    val b = Input(UInt(1.W))
    val x = Output(UInt(1.W))
    val y = Output(UInt(1.W))
}

class TopModule_5 extends RawModule {
    val ports = new TopModuleIO_5
    val a = IO(ports.a)
    val b = IO(ports.b)
    val x = IO(ports.x)
    val y = IO(ports.y)

    x := a ^ b
    y := ~(a^b)
}

val vlog = (new ChiselStage).emitVerilog(new TopModule_5)
println(vlog)
```

## Declaring Wires

refer to https://hdlbits.01xz.net/wiki/Wire_decl

```scala
class TopModuleIO_6 extends Bundle {
    val a = Input(UInt(1.W))
    val b = Input(UInt(1.W))
    val c = Input(UInt(1.W))
    val d = Input(UInt(1.W))
    val out = Output(UInt(1.W))
    val out_n = Output(UInt(1.W))
}

class TopModule_6 extends RawModule {
    val ports = new TopModuleIO_6 
    val a = IO(ports.a)
    val b = IO(ports.b)
    val c = IO(ports.c)
    val d = IO(ports.d)
    val out = IO(ports.out)
    val out_n = IO(ports.out_n)

    out := (a&b) | (c&d)
    out_n := ~out
}

val vlog = (new ChiselStage).emitVerilog(new TopModule_6)
println(vlog)    
```

## 7458

refer to https://hdlbits.01xz.net/wiki/7458

```scala
class TopModuleIO_7 extends Bundle {
    val p1a = Input(UInt(1.W))
    val p1b = Input(UInt(1.W))
    val p1c = Input(UInt(1.W))
    val p1d = Input(UInt(1.W))
    val p1e = Input(UInt(1.W))
    val p1f = Input(UInt(1.W))

    val p2a = Input(UInt(1.W))
    val p2b = Input(UInt(1.W))
    val p2c = Input(UInt(1.W))
    val p2d = Input(UInt(1.W))
    
    val p1y = Output(UInt(1.W))
    val p2y = Output(UInt(1.W))
}

class TopModule_7 extends RawModule {
    val ports = IO(new TopModuleIO_7)

    val w1 = ports.p1a & ports.p1b & ports.p1c
    val w2 = ports.p1d & ports.p1e & ports.p1f 
    val w3 = ports.p2a & ports.p2b 
    val w4 = ports.p2c & ports.p2d

    ports.p1y := w1 | w2
    ports.p2y := w3 | w4
}

val vlog = (new ChiselStage).emitVerilog(new TopModule_7)
println(vlog)
```
