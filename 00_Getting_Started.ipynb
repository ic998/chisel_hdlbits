# 00. Getting Started

## Setup Environment

```scala
// val scalaVersion:="2.12.10"

import $ivy.`edu.berkeley.cs::chisel3:3.4.4`

import chisel3._
import chisel3.stage.ChiselStage
```

## Exercise 1
Build a circuit with no inputs and one output. That output should always drive 1 (or logic high). 

```scala
class TopModuleIO extends Bundle {
    val one = Output(UInt(1.W))
}

class TopModule extends RawModule {
    val bundle = new TopModuleIO

    val one = IO(bundle.one)
    one := 1.U 
}

val vlog = (new ChiselStage).emitVerilog(new TopModule)
println(vlog)
//chisel3.Driver.execute(Array("--target-dir", "generated"), () => new TopModule_00_1)
```

## Exercise 2
Build a circuit with no inputs and one output. That output is driven to a constans 5.

```scala
class TopModuleIO extends Bundle {
    val zero = Output(UInt(4.W))
}

class TopModule extends RawModule {
    val bundle = new TopModuleIO

    val zero = IO(bundle.zero)
    zero := 5.U(4.W)
}

val vlog = (new ChiselStage).emitVerilog(new TopModule)
println(vlog)
//chisel3.Driver.execute(Array("--target-dir", "generated"), () => new TopModule_00_2)
```
